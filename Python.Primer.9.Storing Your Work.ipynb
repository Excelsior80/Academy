# Storing your work

Its all very well writing great code or processing data for whatever reason, but unless it is stored somewhere it's going to go away when the power is turned off. Luckily for us Python offers a suite of tools that enable the storage of data, in this section we will only focus on store to file and get from file, there are of course many ways in which this could be done, databases for example, python also supports interactions with all major data repositories. 

## File management 
For our use case, we will use file management, Python in this context will use access to the underlying hardware through OS abstraction. This means you need not be concerned with the activities under the hood required to get and store binary data to devices, for Python you only have to specify that you are loakding something or saving something. 

### File open
open is a function that is used to open a file, it can be opened in different modes, depending on need. It can be opened as read only, disallowing any modification to it, it can be opened in write mode which will open the file or create it if it exists. Note though that the write mode will clobber a file, in other words it will overwrite whatever is there, this may not be desirable, it can be opened in append mode where the file is opened or created and any new data is added to the end of the file. 

There are a couple of formats that are of use, the file may be text or it may be binary.

run the next cell to get more information on what open is 

```python
print(help(open))
```

### Additional notes on open
File open returns a file object which has it's own methods and attributes for getting information and manipulating the content. Review the next cell:

```python
file_object = open("dictionary.csv", "r")
help(file_object)
file_object.close()
```

Note the public facing methods, read and readline, lets review what happens with readline: 

```python
file_object = open("dictionary.csv", "r")
for line in file_object.readlines():
    print(line)
file_object.close()
```
