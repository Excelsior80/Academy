# Python constructs

Python constructs are an extension of thetypes and operators where we move into the articulation of those in different ways to implement differnet solutions to problems. 

## Programming constructs of control
Programming constructs of control are where statements, functions, or blocks of code are executed in a way that supports what we are trying to achieve. 

The fundamentals are : 
- sequence : one thing after another thing 
- decision : one thing or another thing depending on a test
- Iteration : Repeating a thing until a condition is met



### Sequence 
In programming, a sequence occurs when a group of statements is executed individually, with none repeated and none missed.  It is the most basic programming construct.


```python
# a sequence of statements
name = input("What is your name : ")
length_of_name = len(name)
print(f"The length of the name {name} is {length_of_name} characters")
```

### Selection 
Selections (or branches) are an essential aspect of Python programming, allowing us to make choices.
The “if” statement is the most well-known selection statement:



```python
# a selection 
result = 0
first_number = int(input("What is the first number "))
second_number = int(input("What is the second number "))
choice  = input("Select + or - :")
if choice == '+':
    result = first_number + second_number
else:
    result = first_number - second_number
    
print(f"{first_number} {choice} {second_number} = {result}")
```

If needed more choices can be added : 

```python
# extended selection 
result = 0
first_number = int(input("What is the first number "))
second_number = int(input("What is the second number "))
choice  = input("Select + or - or * :")
if choice == '+':
    result = first_number + second_number
elif choice == '-':
    result = first_number - second_number
elif choice == '*':
    result = first_number * second_number
    
print(f"{first_number} {choice} {second_number} = {result}")
```

### Iteration

An iteration (or loop) allows the programmer to repeat a chosen block of Python code 0 or more times while a given condition is true.  When the controlling condition becomes false the loop is terminated and normal flow is resumed.
In traditional programming, iterations can be pre- or post-conditioned.



```python
#simple loop 
counter = 0
while counter < 5:
    print(f"Counter is {counter}")
    counter += 1
print("Loop finished")
```

## Mutable and immutable
The concept of mutability is something in software that states if something can or cannot be altered.

### Mutable 
If it is mutable, it can be changed

### Immutable 
If it is immutable it cannot be changed
   

```python
# a mutable object 
my_list = ['rice', 'water', 'chicken']
print(my_list)
my_list[1] = 'white wine'
print(my_list)
```

```python
# an immutable object
my_string = "chocolate"
my_string[2] = 'x'
```

As can be seen, an attempt to change an element of a string is rejected.

## Comprehensions
Comprehensions is indeed one of the very useful features in Python, it's a pythonic way of delveloping new data structures based off existing ones. 

Taking the data structure List, we can apply comprehension to it, we can get a list out of a loop, so take something that has the facility to be iterated over and make a new list out of it.

This is a simple example 

```python
#Make a list of numbers between 1 and 20
number_range = range(20) # This is a function that gives you a range of numeric values between 0 and 10
print(number_range)

# you need to be aware of something if you are using range : 
for number in number_range:
    print(number, ' ', end='')
```

```python
#as you can see it starts from zero, that doesnt help, we could just add one to the number in the sequence but, not the best way
number_range = range(1,21)

for number in number_range:
    print(number, ' ', end='')
```

```python
# Great that's what we wanted, lets build a list out of it
number_list = list(number_range)
print (number_list)
```

```python
# great we did a cast to get it into list , there is another way though, using append
looped_numbers = []
for number in range(1,21):
    looped_numbers.append(number)
print(looped_numbers)
```

```python
# ok that worked too, lets try comprehension though 
comprehension_numbers = [number for number in range(1,21)]
print (comprehension_numbers)
```

```python
#great, that last way is cleaner, but it is not all that comprehension does, run the following cells
```

```python
# apply a filter to find only even numbers 
even_numbers = [number for number in comprehension_numbers if number %2 == 0] # the % here is modulo, gives the remainder of a division of two numbers

print(even_numbers)
```

```python
# this time lets add some additional functionality to the same process 
def get_integer_ratio(number_to_examine):
    return (number_to_examine.as_integer_ratio())

ratios = [get_integer_ratio(number) for number in comprehension_numbers if number %2 ==0] # same as before, get the number if it is even but this time go get it's ratio

print(ratios)
```

## Dictionary comprehension
As for list comprehension we can do the same for dictionary : 


```python
my_cats = ["Ron", "Phil", "Jess", "Frank"]
my_cats_lengths = {name: len(name) for name in my_cats}
print(my_cats_lengths)
```
